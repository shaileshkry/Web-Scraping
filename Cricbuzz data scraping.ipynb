### Finding all

```python
most_runs=[]
j=0
for i in soup.find_all('tr',class_='cb-srs-stats-tr'):
    if j==0:
        col=i.text.split()
        j+=1
    #print(i.text.split())
    else:
        most_runs.append(i.text.split())
print(col)
pd.DataFrame(most_runs)
```

```python
soup.find_all('td',class_="cb-srs-stats-td")
```

```python
j=0
a=[]
most_runs=[]
for i in soup.find_all('td',class_="cb-srs-stats-td"):
    if j<8:
        a.append(i.text)
        j+=1
    if j==8:
        #print(a)
        most_runs.append(a)
        j=0
        a=[]
        
df=pd.DataFrame(most_runs)
df
```
