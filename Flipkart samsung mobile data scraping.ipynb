```python
import requests as rq
from bs4 import BeautifulSoup as bsp
import pandas as pd
```

```python
l=[]
for i in range(0,20):
    l1=[]
    model=soup.find_all(class_='_4rR01T')[i].text
    l1.append(model)
    details=soup.find_all(class_='_1xgFaf')[i].text
    l1.append(details)
    price=soup.find_all(class_='_30jeq3 _1_WHN1')[i].text
    l1.append(price)
    rating=soup.find_all(class_='_3LWZlK')[i].text
    l1.append(rating)
    rewiews=soup.find_all(class_='_2_R_DZ')[i].text
    l1.append(rewiews.split('&'))
    l.append(l1)
```

```python
pd.DataFrame(l)
```
