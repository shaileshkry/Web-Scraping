```python
import requests as rq
from bs4 import BeautifulSoup as bsp
import re
```

```python
link='https://www.sarkariresult.com/latestjob/'
```

```python
res=rq.get(link)
res
```

```python
html=res.text
```

```python
soup=bsp(html,'html.parser')
```

```python
soup.find('li')
```

```python
b=str(soup.find(id='post'))
pattern='href=(.*)target'
link=re.findall(pattern,b)[0][1:-2]
link
```

```python
res=rq.get(link)
html=res.text
soup=bsp(html,'html.parser')
soup
```

```python
soup.find('title').text[2:]
```

```python
j=0
title=soup.find('title').text[2:]
pdf=title +"\n"
print(soup.find('title').text[2:])

for i in soup.find_all('li') or soup.find_all('h1') or soup.find_all('h2') or soup.find_all('b'):
    if j in range(10,30):
        print(i.text)
        pdf+=str(i.text)+'\n'
    j+=1
```

```python
print(pdf)
```

```python
from gtts import gTTS
```

```python
tts=gTTS(text=pdf,lang='en',slow=False)
```

```python
tts.save(title + '.mp3')
```
